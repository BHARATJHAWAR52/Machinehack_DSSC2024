```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
read=pd.read_csv("D:/machine_hack/Partcipants_Dataset_DSSC_2024/train.csv")
read.head()
```

```python
label_name={1:"Human Necessities",2:"Transporting",3:"Chemistry",4:"Textiles",5:"Constructions",6:"Mechanical",7:"Physics",8:"Electricity",9:"technology"}
read['Label'] = read['Label'].replace(label_name)
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data=read, x='Label', discrete=True)
plt.xlabel('Patent_class')
plt.ylabel('Count')
plt.title('Patent')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
label_counts = read['Label'].value_counts()
label_counts
```

```python
# Group DataFrame by 'Label'
grouped = read.groupby('Label')

# Generate word cloud for each group
for label, group in grouped:
    text = ' '.join(group['Content'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Plot the word cloud
    plt.figure(figsize=(6, 3))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for Label: {label}')
    plt.axis('off')
    plt.show()
```
